Identity Matrix:
An identity matrix is a special type of square matrix in linear algebra where the diagonal elements along the main diagonal are
equal to 1 and off-diagonal elements taht is all the elements outside the main diagonal (i.e., above and below the diagonal) 
are equal to 0.

In mathematical notation, an identity matrix is represented as follows:
For a 2x2 identity matrix (I2):
I2 = | 1  0 |
     | 0  1 |

For a 3x3 identity matrix (I3):
I3 = | 1  0  0 |
     | 0  1  0 |
     | 0  0  1 |

When we multiply any matrix by an identity matrix of the the appropriate size, it does not change the original matrix. 
In other words, the result of the multiplication is the same as the original matrix. 
Mathematically, if A is any matrix of appropriate dimensions, then:
A x I = A
I x A = A
In the context of matrix operations, multiplying by the identity matrix is equivalent to applying no transformation or 
change to the original matrix. 







```python
# Creating an identity matrix of the same size as img_average_channel
identity_matrix = np.identity(img_average_channel.shape[0])

#Printing a random row out of 'identity_matrix'
# Generate a random integer within the valid range of row indices
random_row_index = np.random.randint(0, identity_matrix.shape[0])

# Print the randomly selected row
random_row = identity_matrix[random_row_index, :]
print("Random Row of Identity Matrix:", random_row)
```

## Part 5

1. Use your array of random row indices to create a new matrix `T` from `identity_matrix`. Overall, `T` has the same elements as `identity_matrix`, but the order of the rows of `T` should be according to the variable `rows`
2. Take a moment to think about what we are doing here. Describe in your own words what matrix `T` is.
3. Sanity check, print the shape of `T` and make sure it is the same size as `identity_matrix` and `img_average_channel`. If you need to,  reshape it to have the same size.


```python
#Using the array of random row indices (rows) to create a new matrix T from identity_matrix
T = identity_matrix[rows]
```

Matrix T is an identity_matrix with its rows shuffled according to the random row indices in the 'rows' array.
It represents a permutation of the rows of the identity matrix.

```python
#Reshaping the size of T if its size is not equal to identity_matrix and printing the shape
if T.shape != identity_matrix.shape:
    T = T.reshape(identity_matrix.shape)  
print("Shape of T:", T.shape)
```

## Part 6

1. Matrix multiply `T` and `img_average_channel` and store the result in `transformed_image`. Check the shape of the result.
2. Plot `transformed_image` and describe the pattern you see. 
3. Think about what each dot product in the matrix multiplication is doing. Describe in your own words what this transformation is doing to the original image. 

```python
#Matrix multiplying T and img_average_channel to get transformed_image
transformed_image = np.dot(T, img_average_channel)

# Checking the shape of the result
print("Shape of transformed_image:", transformed_image.shape)

#Ploting transformed_image
plt.imshow(transformed_image)
plt.title("Transformed Image")
plt.axis('off')
plt.show()
```

When we plot transformed_image, we can observe a pattern that resembles a permutation of the rows of the original img_average_channel. The pattern is scrambled compared to the original image as shown above.

Each dot product in the matrix multiplication is essentially taking a linear combination of the rows of img_average_channel using the row weights specified by the rows of T. Since T is a permutation of the rows of the identity matrix, it shuffles and reorders the rows of img_average_channel. This transformation is rearranging the rows of the original image based on a specific permutation, resulting in the pattern we observe in the transformed_image.

## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python
#Checking if T is invertible by calculating its determinant and rank
det_T = np.linalg.det(T)
rank_T = np.linalg.matrix_rank(T)

#Checking if the determinant is nonzero and if the rank equals the size of T
is_invertible = (det_T != 0) and (rank_T == T.shape[0])
```

```python
if is_invertible==True:
   print("T is invertible")
else:
    print("t is not invertible")
```

```python
if is_invertible:
    Tinv = np.linalg.inv(T)
else:
    print("Since T is not invertible,we cannot calculate inverse of T")
```

```python
# Perform the inverse operation
retrieved_image = np.dot(Tinv, transformed_image)

# Plot the retrieved image
plt.imshow(retrieved_image)
plt.title("Retrieved Original Image")
plt.axis('off')
plt.show()
```

## Extra Points

* Now can you perform the scrambling on __both the columns and rows__ of the original image?  Write the code and plot the scrambled image. Then invert the operations and show that you can recover the original image. 

```python
print("Original Image Shape:", image.shape)

num_rows, num_columns, num_channels = image.shape

rows = np.arange(num_rows)
columns = np.arange(num_columns)

np.random.shuffle(rows)
np.random.shuffle(columns)

scrambled_image = image[rows][:, columns]

print("Shape of scrambled_image:", scrambled_image.shape)

plt.imshow(scrambled_image.astype(np.uint8))
plt.title("Scrambled Image")
plt.show()

inverse_rows = np.argsort(rows)
inverse_columns = np.argsort(columns)

recovered_image = scrambled_image[inverse_rows][:, inverse_columns]

print("Shape of recovered_image:", recovered_image.shape)

plt.imshow(recovered_image.astype(np.uint8))
plt.title("Recovered Original Image")
plt.show()
```
